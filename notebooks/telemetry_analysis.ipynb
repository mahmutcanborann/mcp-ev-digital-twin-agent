```python
import pandas as pd

telemetry_df = pd.read_csv(
    "../models/electric-vehicle-telemetry-dataset.csv"
)

print(telemetry_df.shape)
telemetry_df.head()
```

```python
telemetry_df[
    [
        "SPD",
        "ACC",
        "BRK",
        "ECO",
        "CUR",
        "VOL"
    ]
].describe()
```

```python
telemetry_df["POWER_W"] = (
    telemetry_df["VOL"] *
    telemetry_df["CUR"]
)

telemetry_df["POWER_KW"] = (
    telemetry_df["POWER_W"] / 1000
)

telemetry_df["POWER_KW"].describe()
```

```python
avg_speed = telemetry_df["SPD"].mean()

high_speed_ratio = (
    telemetry_df["SPD"] > 100
).mean() * 100

avg_acc = telemetry_df["ACC"].mean()

hard_brake_ratio = (
    telemetry_df["BRK"] > 10
).mean() * 100

print("Average Speed:", avg_speed)
print("High Speed Ratio:", high_speed_ratio)
print("Average ACC:", avg_acc)
print("Hard Brake Ratio:", hard_brake_ratio)
```

```python
driving_score = 100

driving_score -= high_speed_ratio * 2
driving_score -= hard_brake_ratio * 3
driving_score -= avg_acc * 0.5

driving_score = max(0, min(100, driving_score))

print("Driving Score:", round(driving_score, 2))
```

```python
if driving_score >= 80:
    style = "Eco"
elif driving_score >= 60:
    style = "Normal"
else:
    style = "Aggressive"

print(style)
```
